```python
import os
import re

# Directory where the extracted dataset is located
dataset_directory = '/content/whisper-finetune/dataset'

# Initialize an empty dictionary to store the mapping
audio_transcription_mapping = {}

# Iterate through the files in the dataset directory
for filename in os.listdir(dataset_directory):
    if filename.endswith('.wav'):  # Check if it's an audio file
        audio_filename = filename
        # Construct the expected transcription filename based on the audio filename
        # Replace 'audio' with 'Transcription for' in the filename
        transcription_filename = re.sub(r'_audio\.wav', '_Transcription for _audio.wav', audio_filename)

        # Read the transcription from the corresponding file (if it exists)
        transcription_filepath = os.path.join(dataset_directory, transcription_filename)
        if os.path.isfile(transcription_filepath):
            with open(transcription_filepath, 'r') as transcription_file:
                transcription = transcription_file.read().strip()
            audio_transcription_mapping[audio_filename] = transcription

# Print the created mapping
print(audio_transcription_mapping)
```

```python
import os
import csv  # Import the csv module

# Directory where the extracted dataset is located
dataset_directory = '/content/whisper-finetune/dataset'

# Directories to save the output files
output_directory = '/content/processed_data'
os.makedirs(output_directory, exist_ok=True)  # Create the directory if it doesn't exist

audio_paths_file = os.path.join(output_directory, 'audio_paths.txt')
text_file = os.path.join(output_directory, 'text.txt')

# Read the TSV file (line_index.tsv)
tsv_file_path = os.path.join(dataset_directory, 'line_index.tsv')

if os.path.isfile(tsv_file_path):
    with open(tsv_file_path, 'r', encoding='utf-8') as tsv_file, \
         open(audio_paths_file, 'w', encoding='utf-8') as audio_paths, \
         open(text_file, 'w', encoding='utf-8') as text:

        tsv_reader = csv.reader(tsv_file, delimiter='\t')

        for row in tsv_reader:
            if len(row) >= 2:
                audio_filename = row[0]  # Assuming the audio filename is in the first column
                transcription = row[1]  # Assuming the transcription is in the second column

                # Write to audio_paths and text files
                audio_file_path = os.path.join(dataset_directory, audio_filename)
                audio_paths.write(f'{audio_filename}\t{audio_file_path}\n')
                text.write(f'{audio_filename}\t{transcription}\n')

                # Print the audio file path and transcription
                print(f'Audio File: {audio_file_path}, Transcription: {transcription}')


```

```python
import os
import csv

# Directory where the extracted dataset is located
dataset_directory = '/content/whisper-finetune/dataset'

# Directories to save the output files
output_directory = '/content/processed_data'
os.makedirs(output_directory, exist_ok=True)  # Create the directory if it doesn't exist

audio_paths_file = os.path.join(output_directory, 'audio_paths.txt')
text_file = os.path.join(output_directory, 'text.txt')

# Read the TSV file (line_index.tsv)
tsv_file_path = os.path.join(dataset_directory, 'line_index.tsv')

if os.path.isfile(tsv_file_path):
    with open(tsv_file_path, 'r', encoding='utf-8') as tsv_file, \
         open(audio_paths_file, 'w', encoding='utf-8') as audio_paths, \
         open(text_file, 'w', encoding='utf-8') as text:

        tsv_reader = csv.reader(tsv_file, delimiter='\t')

        for row in tsv_reader:
            if len(row) >= 2:
                # Assuming the audio filename is the unique identifier
                unique_id = row[0]
                transcription = row[1]

                # Construct the absolute path to the audio file
                audio_file_path = os.path.join(dataset_directory, unique_id)

                # Write to audio_paths and text files
                audio_paths.write(f'{unique_id} {audio_file_path}\n')
                text.write(f'{unique_id} {transcription}\n')

                # Print the formatted data for verification
                print(f'{unique_id} {audio_file_path}')
                print(f'{unique_id} {transcription}')

```
